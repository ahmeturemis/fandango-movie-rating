```python
import matplotlib.pyplot as plt
import pandas as pd
movie_reviews = pd.read_csv("fandango_scores.csv")

fig = plt.figure(figsize=(5,12))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)

ax1.set_xlim(0,5.0)
ax2.set_xlim(0,5.0)
ax3.set_xlim(0,5.0)
ax4.set_xlim(0,5.0)

movie_reviews["RT_user_norm"].hist(ax=ax1)
movie_reviews["Metacritic_user_nom"].hist(ax=ax2)
movie_reviews["Fandango_Ratingvalue"].hist(ax=ax3)
movie_reviews["IMDB_norm"].hist(ax=ax4)
plt.show()
```

```python
def calc_mean(series):
    val = series.values
    return sum(val)/len(val)
col = ["RT_user_norm","Metacritic_user_nom","Fandango_Ratingvalue","IMDB_norm"]
user_reviews = movie_reviews[col]
user_reviews_mean = user_reviews.apply(calc_mean)
rt_mean = user_reviews_mean[0]
mc_mean = user_reviews_mean[1]
fg_mean = user_reviews_mean[2]
id_mean = user_reviews_mean[3]
print("RT_user_norm",rt_mean)
print("Metacritic_user_nom",mc_mean)
print("Fandango_Ratingvalue",fg_mean)
print("IMDB_norm",id_mean)
```

```python
def calc_mean(series):
    vals = series.values
    mean = sum(vals) / len(vals)
    return mean
def calc_variance(obj):
    mean = sum(obj)/len(obj)
    count = 0
    for i in obj:
        dif = i - mean
        dif_sq = dif**2
        count += dif_sq
    return count / len(obj)
rt_var = calc_variance(movie_reviews["RT_user_norm"])
rt_stdev = calc_variance(movie_reviews["RT_user_norm"]) ** (1/2)
mc_var = calc_variance(movie_reviews["Metacritic_user_nom"])     
mc_stdev = calc_variance(movie_reviews["Metacritic_user_nom"]) **(1/2)
fg_var =  calc_variance(movie_reviews["Fandango_Ratingvalue"]) 
fg_stdev =  calc_variance(movie_reviews["Fandango_Ratingvalue"]) ** (1/2)
id_var =  calc_variance(movie_reviews["IMDB_norm"]) 
id_stdev =  calc_variance(movie_reviews["IMDB_norm"]) **(1/2)

print("RT_user_norm (varyans)",rt_var)
print("RT_user_norm (standard deviation)",rt_stdev)
print("Metacritic_user_nom (varyans)",mc_var)
print("Metacritic_user_nom (standard deviation)",mc_stdev)
print("Fandango_Ratingvalue (varyans)",fg_var)
print("Fandango_Ratingvalue (standard deviation)",fg_stdev)
print("IMDB_norm (varyans)",id_var)
print("IMDB_norm (standard deviation)",id_stdev)
```

```python
def calc_mean(series):
    vals = series.values
    mean = sum(vals) / len(vals)
    return mean
def calc_covariance(series1,series2):
    mean1 = calc_mean(series1)
    mean2 = calc_mean(series2)
    count = 0
    dif1 = [i - mean1 for i in series1]
    dif2 = [i - mean2 for i in series2]
    for i in range(0,len(series1)):
        count += dif1[i]*dif2[i]   
    return count/len(series1)
rt_fg_covar = calc_covariance(movie_reviews["RT_user_norm"],movie_reviews["Fandango_Ratingvalue"])
mc_fg_covar = calc_covariance(movie_reviews["Metacritic_user_nom"],movie_reviews["Fandango_Ratingvalue"])
id_fg_covar = calc_covariance(movie_reviews["IMDB_norm"],movie_reviews["Fandango_Ratingvalue"])
print("Covariance between the RT_user_norm and Fandango_Ratingvalue :",rt_fg_covar)
print("Covariance between the Metacritic_user_nom and Fandango_Ratingvalue :",mc_fg_covar)
print("Covariance between the IMDB_norm and Fandango_Ratingvalue :",id_fg_covar)
```

```python
def calc_correlation(series1, series2):
    stdev = (calc_variance(series1)**(1/2))*(calc_variance(series2)**(1/2))
    return calc_covariance(series1,series2)/stdev
rt_fg_corr = calc_correlation(movie_reviews["RT_user_norm"], movie_reviews["Fandango_Ratingvalue"])
mc_fg_corr = calc_correlation(movie_reviews["Metacritic_user_nom"], movie_reviews["Fandango_Ratingvalue"])
id_fg_corr = calc_correlation(movie_reviews["IMDB_norm"], movie_reviews["Fandango_Ratingvalue"])
print("Correlation between the RT_user_norm and Fandango_Ratingvalue :",rt_fg_corr)
print("Correlation between the Metacritic_user_nom and Fandango_Ratingvalue :",mc_fg_corr)
print("Correlation between the IMDB_norm and Fandango_Ratingvalue :",id_fg_corr)
```
